```python
import tensorflow as tf
```

```python
neural_network = tf.keras.models.Sequential()
neural_network.add(tf.keras.layers.Dense(units=2, activation='relu', input_shape=(3,)))
neural_network.add(tf.keras.layers.Dense(units=2, activation='relu'))
neural_network.add(tf.keras.layers.Dense(units=2, activation='relu'))
neural_network.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

```python
neural_network.summary()
```

```python
neural_network.compile(optimizer='Adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
neural_network.fit(x_credit_treinamento, y_credit_treinamento, epochs=2500)
```

```python
y_pred = neural_network.predict(x_credit_teste)
```

```python
y_pred = y_pred >= 0.5
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_credit_teste, y_pred)
plt.figure(figsize=(12,7))
sns.heatmap(cm, annot=True, cmap='viridis', fmt='g');
```

```python
print(classification_report(y_credit_teste, y_pred))
```
