```python
from sklearn.model_selection import GridSearchCV
```

```python
x_credit = np.concatenate((x_credito_treinamento, x_credito_teste), axis=0)
y_credit = np.concatenate((y_credito_treinamento, y_credito_teste), axis=0)
print(x_credit.shape)
x_credit.shape
```

```python
parameters = {'criterion': ['gini', 'entropy'],
              'splitter': ['best', 'random'],
              'min_samples_split': [2, 5, 10],
              'min_samples_leaf': [1, 5, 10]}
```

```python
grid_search = GridSearchCV(estimator=DecisionTreeClassifier(), param_grid=parameters)
grid_search.fit(x_credit, y_credit)
best_parameters = grid_search.best_params_
best_score = grid_search.best_score_
print(best_parameters)
print(best_score)
```

```python
def prepare_tree_parameters(parameters):
    equal_or_comma = ','
    converted_parameters = ''
    for i in range(len(parameters)):
        if equal_or_comma == ',':
            if parameters[i] == "'":
                converted_parameters += ''
            elif parameters[i] == ":":
                converted_parameters += ' ='
                equal_or_comma = '='
            else:
                converted_parameters += parameters[i]
        elif equal_or_comma == '=': 
            if parameters[i] == "'":
                converted_parameters += parameters[i]
            elif parameters[i] == ",":
                converted_parameters += ','
                equal_or_comma = ','
            else:
                converted_parameters += parameters[i]
                
            
    return converted_parameters
```

```python
parameters = prepare_tree_parameters("'criterion': 'entropy', 'min_samples_leaf': 1, 'min_samples_split': 5, 'splitter': 'best'")
print(parameters)
```

```python
from sklearn.model_selection import cross_val_score, KFold
results = []

for i in range(30): 
    
    kfold = KFold(n_splits=10, shuffle=True, random_state=i)
    tool = DecisionTreeClassifier(criterion = 'entropy', min_samples_leaf = 1, min_samples_split = 5, splitter = 'best')
    scores = cross_val_score(tool, x_credit, y_credit, cv = kfold)
    # print(scores)
    results.append(scores.mean())
results
```
