```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle
from sklearn.tree import DecisionTreeClassifier
```

```python
with open('./Machine Learning e Data Science com Python de A Ã  Z/Bases de dados/risco_credito.pkl', 'rb') as f:
    x_risco_credito, y_risco_credito = pickle.load(f)
```

```python
arvore_risco_credito = DecisionTreeClassifier(criterion='entropy')
arvore_risco_credito.fit(x_risco_credito, y_risco_credito)
```

```python
arvore_risco_credito.feature_importances_
```

```python
from sklearn import tree
previsores = ['histÃ³ria de crÃ©dito', 'dÃ­vida', 'garantias', 'renda']
figura, eixos = plt.subplots(nrows=1, ncols=1, figsize=(15, 15))
tree.plot_tree(arvore_risco_credito, feature_names=previsores, class_names=arvore_risco_credito.classes_, filled=True);
```
